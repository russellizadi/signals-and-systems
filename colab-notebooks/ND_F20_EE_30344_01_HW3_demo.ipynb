```
import matplotlib.pyplot as plt
import numpy as np
import scipy.io
from scipy import signal
from IPython.display import Audio
```

```
# upload the lineup.mat through the "files" section on the left

# load 
data = scipy.io.loadmat('lineup.mat')
y = data['y'][:, 0]
fs = 8192
print(f"duration = {len(y)/fs} (s)")

# plot
n = np.arange(0, len(y))
plt.plot(n, y)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("y")
plt.show()

# play
Audio(y,rate=fs,autoplay=True)
```

```
# a
def dirac(n):
    return np.where(n == 0, 1.0, 0.0)

a = .5
N = 1000

# change
ne = np.arange(0, 1001)
he = dirac(ne-1)

plt.plot(ne, he)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("he")
plt.show()
```

```
# c
# change
ner = np.arange(4001)
d = dirac(ner)

nom = [1, 0]
den = [2, 4]

her = signal.lfilter(den, nom, d)

plt.plot(ner, her)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("her")
plt.show()
```

```
# d
z = signal.lfilter(den, nom, y)

plt.plot(n, z)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("z")
plt.show()
Audio(z,rate=fs,autoplay=True)
```

```
# e
hoa = np.convolve(he, her)
noa = np.arange(0, len(hoa))
plt.plot(noa, hoa)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("hoa")
plt.show()
```
